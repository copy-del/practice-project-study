```python
# [1] 라이브러리 불러오기
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder

# [2] 데이터 불러오기
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

# 타깃값 저장
y = train['Survived']
train.drop(['Survived'], axis=1, inplace=True)

# train + test 합치기
data = pd.concat([train, test], sort=False)

# [3] 피처 엔지니어링

# ① Title 추출
data['Title'] = data['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
data['Title'] = data['Title'].replace(['Lady', 'Countess', 'Capt', 'Col', 'Don', 
                                       'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')
data['Title'] = data['Title'].replace('Mlle', 'Miss')
data['Title'] = data['Title'].replace('Ms', 'Miss')
data['Title'] = data['Title'].replace('Mme', 'Mrs')

# ② 가족 크기
data['FamilySize'] = data['SibSp'] + data['Parch'] + 1
data['IsAlone'] = (data['FamilySize'] == 1).astype(int)

# ③ Age 결측치: Title 기준 평균으로 채움
title_ages = data.groupby('Title')['Age'].transform('mean')
data['Age'] = data['Age'].fillna(title_ages)

# ④ Fare 결측치: 중앙값
data['Fare'] = data['Fare'].fillna(data['Fare'].median())

# ⑤ Embarked 결측치: 최빈값
data['Embarked'] = data['Embarked'].fillna(data['Embarked'].mode()[0])

# [4] 불필요한 컬럼 제거
data.drop(['Name', 'Ticket', 'Cabin', 'PassengerId'], axis=1, inplace=True)

# [5] 범주형 인코딩
for col in ['Sex', 'Embarked', 'Title']:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])

# [6] train/test 다시 분할
X_train = data[:len(y)]
X_test = data[len(y):]

# [7] 모델 학습 및 평가
clf = RandomForestClassifier(n_estimators=100, random_state=42)
score = cross_val_score(clf, X_train, y, cv=5).mean()

print(f'교차검증 정확도: {score:.4f}')  # 목표: 0.8000 이상
```

```python
# [1] 학습/검증 데이터 분할
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier

X_tr, X_val, y_tr, y_val = train_test_split(X_train, y, test_size=0.2, random_state=42)

# [2] RandomForest 모델 튜닝 및 학습
rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=6,
    min_samples_split=4,
    min_samples_leaf=2,
    random_state=42
)
rf.fit(X_tr, y_tr)

# [3] 검증 정확도 출력
val_pred = rf.predict(X_val)
val_acc = accuracy_score(y_val, val_pred)
print(f'검증 정확도: {val_acc:.4f}')

# [4] test.csv 예측 및 캐글 제출 파일 저장
final_pred = rf.predict(X_test)
submission = test[['PassengerId']].copy()
submission['Survived'] = final_pred
submission.to_csv('titanic_submission_rf_tuned.csv', index=False)
```

```python
# [1] 라이브러리
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.preprocessing import LabelEncoder

# [2] 데이터 로드
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
y = train['Survived']
train.drop(['Survived'], axis=1, inplace=True)
data = pd.concat([train, test], sort=False)

# [3] 피처 엔지니어링
# ① Title 추출 및 간단히 줄이기
data['Title'] = data['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
data['Title'] = data['Title'].replace(['Mlle', 'Ms'], 'Miss')
data['Title'] = data['Title'].replace('Mme', 'Mrs')
data['Title'] = data['Title'].replace(['Lady', 'Countess', 'Capt', 'Col', 'Don', 'Dr',
                                       'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')
# 최종 title 목록: Mr, Miss, Mrs, Master, Rare

# ② 가족 관련 변수
data['FamilySize'] = data['SibSp'] + data['Parch'] + 1
data['IsAlone'] = (data['FamilySize'] == 1).astype(int)

# ③ 결측치 처리
data['Age'] = data['Age'].fillna(data.groupby('Title')['Age'].transform('mean'))
data['Fare'] = data['Fare'].fillna(data['Fare'].median())
data['Embarked'] = data['Embarked'].fillna(data['Embarked'].mode()[0])

# ④ Age, Fare 구간화
data['AgeBin'] = pd.cut(data['Age'], bins=[0, 16, 32, 48, 64, 100], labels=False)
data['FareBin'] = pd.qcut(data['Fare'], 4, labels=False)

# [4] 불필요한 변수 제거
data.drop(['Name', 'Ticket', 'Cabin', 'PassengerId', 'Age', 'Fare'], axis=1, inplace=True)

# [5] 범주형 변수 인코딩
for col in ['Sex', 'Embarked', 'Title']:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])

# [6] 학습용 / 테스트용 분할
X_train = data[:len(y)]
X_test = data[len(y):]

# [7] Stratified K-Fold 교차검증
rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=6,
    min_samples_split=4,
    min_samples_leaf=2,
    random_state=42
)
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_score = cross_val_score(rf, X_train, y, cv=kfold, scoring='accuracy')

print(f'5-Fold 교차검증 평균 정확도: {cv_score.mean():.4f}')
```

```python
# [8] 전체 학습 후 test 데이터 예측
rf.fit(X_train, y)
final_pred = rf.predict(X_test)

# [9] 캐글 제출용 파일 생성
submission = test[['PassengerId']].copy()
submission['Survived'] = final_pred
submission.to_csv('titanic_submission_rf_kfold.csv', index=False)
```

```python
# [1] 라이브러리 불러오기
import pandas as pd
import numpy as np
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.preprocessing import LabelEncoder
from xgboost import XGBClassifier

# [2] 데이터 불러오기
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

# 타겟 분리
y = train['Survived']
train.drop(['Survived'], axis=1, inplace=True)

# 합치기
data = pd.concat([train, test], sort=False).reset_index(drop=True)

# [3] Title 추출 및 간소화
data['Title'] = data['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
data['Title'] = data['Title'].replace(['Mlle', 'Ms'], 'Miss')
data['Title'] = data['Title'].replace('Mme', 'Mrs')
data['Title'] = data['Title'].replace(
    ['Lady', 'Countess', 'Capt', 'Col', 'Don', 'Dr', 'Major',
     'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare'
)

# [4] 가족 관련 변수 생성
data['FamilySize'] = data['SibSp'] + data['Parch'] + 1
data['IsAlone'] = (data['FamilySize'] == 1).astype(int)

# [5] 결측치 처리
data['Age'] = data['Age'].fillna(data.groupby('Title')['Age'].transform('mean'))
data['Fare'] = data['Fare'].fillna(data['Fare'].median())
data['Embarked'] = data['Embarked'].fillna(data['Embarked'].mode()[0])

# [6] Cabin 전처리 → 앞 글자 추출
data['CabinType'] = data['Cabin'].astype(str).str[0]
data['CabinType'] = data['CabinType'].replace('n', 'U')  # NaN 대체

# [7] 수치형 변수 구간화
data['AgeBin'] = pd.cut(data['Age'], bins=[0, 16, 32, 48, 64, 100], labels=False)
data['FareBin'] = pd.qcut(data['Fare'], 4, labels=False)

# [8] 범주형 변수 인코딩
for col in ['Sex', 'Embarked', 'Title', 'CabinType']:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])

# [9] 훈련/테스트 나누기
X_full = data[:len(y)].copy()
X_test = data[len(y):].copy()

# [10] 사용할 피처 선택
features = [
    'Pclass', 'Sex', 'Title',
    'Age', 'Fare', 'CabinType', 'SibSp', 'Parch',
    'FamilySize', 'IsAlone'
]

X_train = X_full[features]
X_test = X_test[features]

# [11] 모델 정의
xgb = XGBClassifier(
    n_estimators=800,
    max_depth=5,
    learning_rate=0.03,
    subsample=0.9,
    colsample_bytree=0.8,
    min_child_weight=1,
    gamma=0,
    use_label_encoder=True,
    eval_metric='logloss',
    random_state=50
)

# [12] 교차검증 (10-Fold)
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
score = cross_val_score(xgb, X_train, y, cv=kfold, scoring='accuracy')
print(f'10-Fold 교차검증 평균 정확도: {score.mean():.4f}')
```
