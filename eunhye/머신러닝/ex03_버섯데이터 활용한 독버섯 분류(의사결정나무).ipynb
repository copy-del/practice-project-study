```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

#의사결정 나무 모델 불러오기 
from sklearn.tree import DecisionTreeClassifier
```

```python
# 데이터 불러오기 
data = pd.read_csv('./머신러닝/data/mushroom.csv')
data
```

```python
data.shape
#8124개의 데이터
#입력특성(22개) + 정답 (1개)
```

```python
#데이터 정보 확인
data.info()

# 1. 결측치 여부 확인 -> 결측이 벗음
# 2. 데이터 타입 확인 -> object(문자열)
# 사이킷런의 알고리즘은 문자열 값을 입력값으로 받지 않음!
```

#### 문제와 정답으로 분리
- 문제 데이터 x (특성, 속성, 피처, 독립변수) : poisonous 제외한 모든 다른 칼럼
- 정답 데이터 y (정답, 라벨, 레이블, 타겟, 종속변수) : poisonous 

```python
# X = data.iloc[ : , 1: ]
# X = data.loc[ : , 'cap-shape': ]

X = data.drop('poisonous', axis = 1)
y = data['poisonous']

X
```

```python
y
```

#### 데이터 전처리
- 문자열 데이터 -> 수치형 (인코딩, encoding)


```python
y.value_counts()
```

```python
# 기술통계량 확인
data.describe()
```

```python
X['cap-shape'].unique()
```

```python
#원 핫 인코딩: 값의 크고 작음에 의미가 없을 때 사용(상하관계가 없을 때)
X_one_hot = pd.get_dummies(X, dtype = 'int64')

X_one_hot
```

### train, test 데이터 분리

```python
from sklearn.preprocessing import LabelEncoder

# 분리 비율 7:3, 고정 규칙(random_state = 1)

X_train, X_test, y_train, y_test = train_test_split(X_one_hot, y, test_size=0.3, random_state=21)
```

```python
# 분리 후 데이터의 크기(형태) 확인
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

#### 모델링
1. 모델객체 생성
2. 모델 학습
3. 모델 평가

```python
#모델 객체 생성/ 하이퍼파라미터 조정 안해봄
tree_model = DecisionTreeClassifier()
```

```python
tree_model.fit(X_train, y_train)
```

```python
#모델 평가
tree_model.score(X_test, y_test)

# 1.0 -> 정확도 100%? 
```

#### 하이퍼파라미터 조절한 모델링

```python
# 모델 객체 생성
tree_model2 = DecisionTreeClassifier(max_depth = 3)
```

```python
tree_model2.fit(X_train, y_train)
```

```python
tree_model2.score(X_test, y_test)
```

```python
#우리가 보유한 데이터는 한정적: 데이터가 많을수록 과대적합을 제어할 수 있는데
#데이터수 변화 없이 어떻게 하면 여러번 검증할 수 있을까?
#교차 검증 : 보유한 데이터를 여러번 활용해서 검증하여 성능을 확인
#일반화 정도를 확인
```

### 특성 선택
- 각 특성들의 중요도를 출력해보자.
- 정답에 얼마만큼의 영향을 미치는 특성인가를 수치로 확인 가능
- 각 특성들의 중요도는 0~1 사이 값을 가진다.
- 모든 특성들의 중요도 합은 1
- tree모델은 특성의 중요도가 가장 높은 값을 상위 질문(root node)으로 배치


```python
#tree model의 특성 중요도 확인
fi = tree_model.feature_importances_
fi
```

```python
pd.set_option('display.max_rows', None)
#판다스 디스플레이 옵션 변경(행 다 보이게)
```

```python
#중요도 DF화
fi_df = pd.DataFrame(fi, index = X_one_hot.columns, columns = ['특성중요도'])
fi_df
```

```python
# 특성 중요도를 확인하기 위해 내림차순으로 정렬
fi_df.sort_values(by='특성중요도', ascending = False)
```
